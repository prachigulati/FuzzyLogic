```python
#program 8
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()


# print(len(x_train)) #60,000 images
# print(len(y_train)) #60,000 labels(digits)
# print(len(x_test)) #10,000 images
# print(len(y_test)) #10,000 labels(digits)
# plt.matshow(x_test[0])
# print(y_test[0])
# x_train.shape #(60000,28,28)


#scale the array(as all the values are between 0 and 255 so scale tem to 0 and 1)
x_train=x_train/255
x_test=x_test/255
# print(x_train[0]) (values will be between 0 and 1)


#flatten the 28 x 28 gird image(2D array) to 1D array
x_train_flattened=x_train.reshape(len(x_train),28*28)
# x_train_flattened.shape #(60000,284)
x_test_flattened=x_test.reshape(len(x_test),28*28)



#create neural netwrok
model = keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation="relu"),
    keras.layers.Dense(10,input_shape=(784,),activation="sigmoid"),
])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train_flattened,y_train, epochs=5)


# evaluate
model.evaluate (x_test_flattened,y_test)

#predict the values
y_predicted=model.predict(x_test_flattened)
# y_predicted[0] #this will give the 10 outputs


#numpy has a function
#use argmax to get the maximum value and it will print the corresponding index vale
# np.argmax(y_predicted[0])


y_predicted_labels=[np.argmax(i) for i in y_predicted]

#confusion matrix
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot= True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')

```

```python
#program 9
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import pandas as pd

# iris is a classic dataset with 150 samples of flower measurements
iris = load_iris()
x = iris.data  #x contains 4 features(sepal length, sepal width, petal length, petal width)
y = iris.target #y contains class labels (0: setosa, 1:versicolor, 2:virginica)

#normalize the data so that all feature values are scaled between 0 and 1
x= x/x.max()

#split the dataset into training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
# 70% training, 30% testing
# random_state=42 ensures reproducibility of results


#supervised learning : neural network model
model = keras.Sequential([
    keras.layers.Dense(16,input_shape=(4,),activation ='relu'),
    keras.layers.Dense(12,activation ='relu'),
    keras.layers.Dense(3, activation='softmax')  #3 output classes for iris
])



model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train,y_train, epochs=250)
model.evaluate(x_test, y_test)


y_pred_probs = model.predict(x_test)#predicts possibilities for each class(softmax output)
y_pred=[np.argmax(i) for i in y_pred_probs] #converts probabilities to class labels(index of high probability)


# print(y_pred[:5])
# y_test[:5]


#confusion matrix
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.title("Confusion Matrix-Iris classification (supervised)")
plt.show()



#unsupervised learning - K means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x)
#applies KMeans clustering to group the data into 3 clusters(without using labes)


iris_df = pd.DataFrame(x, columns=iris.feature_names) #converts the data to a pandas dataframe
iris_df['cluster']=kmeans.labels_ #adds a new column 'cluster' with the predicted cluster label for each sample


plt.figure(figsize=(8,6))
plt.scatter(iris_df.iloc[:,0], iris_df.iloc[:,1], c=iris_df['cluster'], cmap='viridis', s=60)
            #first feature sepal length, second feature sepal width
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red', marker='x', label='centroids')
#kmeans.cluster_centers_: coordinates of the cluster centers
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("K-Means Clustering on Iris Dataset (Unsupervised)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#program 7
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import seaborn as sns

# XOR input and output
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0], [1], [1], [0]])

#define neural network model
model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x,y,epochs=1000)

predictions= model.predict(x)
print("Predictions:",np.round(predictions).astype(int))

def plot_decision_boundary(model, x, y):
  #define the plotting grid area
  x_min, x_max=x[:,0].min() - 0.5, x[:,0].max() +0.5
  y_min, y_max=x[:,1].min() - 0.5, x[:,1].max() +0.5

  #generate a grid of points(100x100)
  xx,yy =np.meshgrid(np.linspace(x_min,x_max,100),np.linspace(y_min,y_max,100))

  #model predictions over the grid
  z = model.predict(np.c_[xx.ravel(),yy.ravel()])
  #.ravel() flattens the grid into 1D
  z = z.reshape(xx.shape)
  plt.figure(figsize=(6, 6))
  plt.contourf(xx, yy, z, alpha=0.8, colors=['blue', 'red'])
  plt.scatter(x[:, 0], x[:, 1], c=y.ravel(), edgecolors='k', cmap='coolwarm')
  plt.xlabel('Feature 1')
  plt.ylabel('Feature 2')
  plt.title('Decision Boundary')
  plt.show()

plot_decision_boundary(model, x, y)

```

```python
#program 6
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)   #makes the random numbers repeatable for consistent output

#generates 10random 2D points centered around(2,2)
x1=np.random.randn(10, 2) + np.array([2,2])
#generates 10 random 2D points centered around(-2,-2)
x2=np.random.randn(10,2) + np.array([-2,-2])


x=np.vstack((x1,x2))   #vertical stack  #shape(20,2)
y=np.hstack((np.ones(10),-np.ones(10))) #horizontal stack  #first 10 have labels 1 next 10 have -1

class Perceptron:
  #constructor sets
  def __init__(self,learning_rate=0.01, epochs=100):
    self.lr=learning_rate   #how much to adjust weights each time
    self.epochs=epochs
    self.weights=None
    self.bias=None

  #training the perceptron
  def fit(self, x, y):
    self.weights=np.zeros(x.shape[1])  #initializes the weight to zero:one weight per feature
    self.bias=0
    for _ in range(self.epochs):
      for i in range(x.shape[0]):
        update = self.lr * (y[i]-self.predict(x[i]))    #update=learningrate x (truelabel - predictedlabel)
        #update weights
        self.weights += update*x[i]
        self.bias += update

  #prediction
  def predict(self, x):
    return np.where(np.dot(x,self.weights)+self.bias >= 0,1,-1)   #if result>=0 then class 1 else -1

perceptron = Perceptron(learning_rate=0.1,epochs=10)
perceptron.fit(x,y)

#plotting the data and decision boundary
plt.scatter(x1[:,0], x1[:,1], color='blue', label='class 1')
plt.scatter(x2[:,0], x2[:,1], color='red', label='class 0')

x_vals=np.linspace(-5,5,100)
y_vals=-(perceptron.weights[0]*x_vals + perceptron.bias)/perceptron.weights[1]
#decision boundary line eqn  w1*x + w2*y + b = 0
#  y=-(w1*x+b)/(w2)


plt.plot(x_vals,y_vals,'g--',label='decision boundary')  #dashed green line
plt.legend()
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.title('Perceptron Decision Boundary')
plt.show()


#test point
test_point = np.array([-1.1,-1.2])
prediction = perceptron.predict(test_point)
print(f"Prediction for test point {test_point}: {'class 1' if prediction ==1 else 'class 0'}")
```

```python
#program 3
import numpy as np
import matplotlib.pyplot as plt

def triangular_membership(x,a,b,c):
  if x<=a or x>=c:
    return 0.0
  elif x==b:
    return 1.0
  elif x<b:
    return (x-a)/(b-a)
  else:
    return (c-x)/(c-b)

def trapezoidal_membership(x,a,b,c,d):
  if x<=a or x>=d:
    return 0
  elif b<=x<=c:
    return 1.0
  elif a<=x<b:
    return (x-a)/(b-a)
  else:
    return (d-x)/(d-c)

def gaussian_membership(x,c,sigma):
  return np.exp(-0.5*((x-c)/sigma)**2)

user_x = float(input("Enter value of x:"))
tri_val = triangular_membership(user_x, 2, 5, 8)
trap_val = trapezoidal_membership(user_x, 2, 4, 6, 8)
gauss_val = gaussian_membership(user_x, 5, 1)


print("Membership Degrees at x = ", user_x)
print("Triangular     :", round(tri_val, 3))
print("Trapezoidal    :", round(trap_val, 3))
print("Gaussian       :", round(gauss_val, 3))

#for plotting the graph
x_vals = np.linspace(0, 10, 1000)
tri_vals=[triangular_membership(x,2,5,8) for x in x_vals]
trap_vals=[trapezoidal_membership(x,2,4,6,8) for x in x_vals]
gauss_vals=[gaussian_membership(x,5,1) for x in x_vals]
plt.plot(x_vals, tri_vals, label='Triangular', color='blue')
plt.plot(x_vals, trap_vals, label='Trapezoidal', color='green')
plt.plot(x_vals, gauss_vals, label='Gaussian', color='red')

#marking the users x
plt.axvline(x=user_x, color='black', linestyle='--', label=f'x = {user_x}')
plt.scatter([user_x], [tri_val], color='blue')
plt.scatter([user_x], [trap_val], color='green')
plt.scatter([user_x], [gauss_val], color='red')

plt.title('Fuzzy Membership Functions')
plt.xlabel('Input Value (x)')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#program 4
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

#defining fuzzy variables input output
temp_diff = ctrl.Antecedent(np.arange(0, 101, 1), 'temp_diff')  #(-30C to 30C)
humidity = ctrl.Antecedent(np.arange(0,101,1), 'humidity')       #0% to 100%
fan_speed = ctrl.Consequent(np.arange(0,101,1),'fan_speed')     #0% to 100%

#define membersip functions
temp_diff['negative'] = fuzz.trimf(temp_diff.universe, [-30, -30, 0])
temp_diff['zero'] = fuzz.trimf(temp_diff.universe, [-5, 0, 5])
temp_diff['positive'] = fuzz.trimf(temp_diff.universe, [0, 30, 30])

humidity['low'] = fuzz.trimf(humidity.universe, [0, 0, 50])
humidity['medium'] = fuzz.trimf(humidity.universe, [30, 50, 70])
humidity['high'] = fuzz.trimf(humidity.universe, [50, 100, 100])

fan_speed['off'] = fuzz.trimf(fan_speed.universe, [0, 0, 25])
fan_speed['low'] = fuzz.trimf(fan_speed.universe, [20, 40, 60])
fan_speed['medium'] = fuzz.trimf(fan_speed.universe, [40, 60, 80])
fan_speed['high'] = fuzz.trimf(fan_speed.universe, [60, 100, 100])


#define fuzzy rules
rule1 = ctrl.Rule(temp_diff['positive'] | humidity['high'], fan_speed['high'])
rule2 = ctrl.Rule(temp_diff['zero'] & humidity['medium'], fan_speed['medium'])
rule3 = ctrl.Rule(temp_diff['negative'] | humidity['low'], fan_speed['off'])

#control and simulate the control system
fan_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
fan = ctrl.ControlSystemSimulation(fan_ctrl)


#user input
temp_input = int(input("Enter the temperature difference (°C): "))
humidity_input = int(input("Enter the humidity level (%): "))

#compute the output
fan.input['temp_diff'] = temp_input
fan.input['humidity'] = humidity_input
fan.compute()

#print the output
print(f"Fan Speed: {fan.output['fan_speed']:.2f}%")


plt.subplot(1,3,1); temp_diff.view()
plt.subplot(1,3,2); humidity.view()
plt.subplot(1,3,3); fan_speed.view()

fan_speed.view(sim=fan)
```

```python
#program 5
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt


#define input output
room_temp = ctrl.Antecedent(np.arange(-30, 31, 1), 'room_temp')
heater_power = ctrl.Consequent(np.arange(0, 101, 1), 'heater_power')

#membership function
room_temp['negative'] = fuzz.trimf(room_temp.universe, [-30, -30, 0])
room_temp['zero'] = fuzz.trimf(room_temp.universe, [-5, 0, 5])
room_temp['positive'] = fuzz.trimf(room_temp.universe, [0, 30, 30])

heater_power['off'] = fuzz.trimf(heater_power.universe, [0, 0, 25])
heater_power['low'] = fuzz.trimf(heater_power.universe, [20, 40, 60])
heater_power['medium'] = fuzz.trimf(heater_power.universe, [40, 60, 80])
heater_power['high'] = fuzz.trimf(heater_power.universe, [60, 100, 100])

rule1 = ctrl.Rule(room_temp['negative'], heater_power['high'])
rule2 = ctrl.Rule(room_temp['zero'], heater_power['medium'])
rule3 = ctrl.Rule(room_temp['positive'], heater_power['off'])

heater_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
heater = ctrl.ControlSystemSimulation(heater_ctrl)

user_input = float(input("Enter the room temperature difference (°C): "))
heater.input['room_temp'] = user_input
heater.compute()


print(f"Heater Power: {heater.output['heater_power']:.2f}%")


#to plot the output curve
temp_range = np.arange(-30, 31, 1)
power_output = []
for temp in temp_range:
    heater.input['room_temp'] = temp
    heater.compute()
    power_output.append(heater.output['heater_power'])


#plotting
plt.figure(figsize=(10, 5))
plt.plot(temp_range, power_output, color='blue', linewidth=2)
plt.title("Heater Power vs Room Temperature Difference")
plt.xlabel("Room Temperature Difference (°C)")
plt.ylabel("Heater Power (%)")
plt.grid(True)
plt.axvline(user_input, color='red', linestyle='--', label='Your Input')
plt.legend()
plt.show()

```
