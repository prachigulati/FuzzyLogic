```python
#Implement a Multi-layer perceptron (MLP) for solving XOR problems with backpropagation.

import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import seaborn as sns

# XOR Dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]]) #input values
y = np.array([[0], [1], [1], [0]])  #output values

# Define MLP model
model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),  # 8 neurons for better learning
    Dense(1, activation='sigmoid')  # Output layer with sigmoid activation
])

#sequential ([..])creates a model where layers are added one after another

# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
#model.compile(...)   prepares the model for training
#optimizer='adam'   uses the adam optimizer, which adjusts learning automatically
#                   for better performancee
#loss='binary_crossentropy'    Binary cross-entropy loss measures how far the model's output
#                              is from the correct answer(used for classification problems)
#matrices=['accuracy']    Measures how often the model's predictions are correct


# Train model with reduced epochs
history = model.fit(X, y, epochs=500, verbose=1)

# Evaluate model
predictions = model.predict(X, batch_size=32)
print("Raw Predictions:", predictions)
print("Rounded Predictions:", np.round(predictions).astype(int))

# Plot loss curve
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Curve')
plt.legend()
plt.show()

# Function to plot decision boundary
def plot_decision_boundary(model, X, y):
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 50), np.linspace(y_min, y_max, 50))  # Reduced grid size

    # Predict class for each point in the meshgrid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()], batch_size=32)
    #.ravel() flattens the grid into 1D array
    # np.c_[xx.ravel(),yy.ravel()]  combines the two flattened arrays into a 2 column array
    # batch_size=32  feeds all grid points into the trained neural network
    #                 the model predicts whether each point belong to 0 or 1
    Z = np.round(Z).reshape(xx.shape)  # Ensure proper shape for contour plot
    # now z is a 2d array where each value represents the predicted class(0or 1) for that grid point


    #plotting the decision boundary
    plt.figure(figsize=(6, 6))   #6x6
    plt.contourf(xx, yy, Z, alpha=0.8, cmap='coolwarm')  # Use a colormap
    #"coolwarm"colormaap=uses red/blue colors to distinguish between classes
    # Blue for class 0 (output = 0)
    # Red for class 1 (output = 1)
    #alpha makes the colors slightly transparent
    plt.scatter(X[:, 0], X[:, 1], c=y.ravel(), edgecolors='k', cmap='coolwarm')
    #plt.scatter(..)plots the original dataset points
    #x[:,0] x values of original data
    #x[:,1] y values of original data
    #c=y.ravel() colors the points based on their actual class(0 or 1)
    #edgecolors='k' black edges for better visibilty
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Decision Boundary')
    plt.show(block=False)  # Non-blocking in interactive environments
    plt.close()  # Ensure it doesn't hang

# Plot decision boundary
plot_decision_boundary(model, X, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Fitness function
def fitness(x):
    return np.sin(x)

def generate_population(size, lower, upper):
    return np.random.uniform(lower, upper, size)

def tournament_selection(pop, fitnesses, k=3):
    selected = []
    for _ in range(len(pop)):
        aspirants_idx = np.random.randint(0, len(pop), k)
        winner = pop[aspirants_idx[np.argmax(fitnesses[aspirants_idx])]]
        selected.append(winner)
    return np.array(selected)

def roulette_wheel_selection(pop, fitnesses):
    fitnesses = np.maximum(fitnesses, 1e-6)
    probs = fitnesses / np.sum(fitnesses)
    return np.random.choice(pop, size=len(pop), p=probs)

def one_point_crossover(p1, p2):
    alpha = random.random()
    return alpha * p1 + (1 - alpha) * p2

def blend_crossover(p1, p2, alpha=0.5):
    gamma = (1 + 2 * alpha) * random.random() - alpha
    return (1 - gamma) * p1 + gamma * p2

def mutate(x, mutation_rate=0.1):
    if random.random() < mutation_rate:
        return x + np.random.normal(0, 0.1)
    return x

def genetic_algorithm(selection_strategy, crossover_strategy, mutation_rate=0.1, crossover_rate=0.9, generations=50, pop_size=50):
    pop = generate_population(pop_size, 0, 2 * np.pi)
    best_scores = []

    for _ in range(generations):
        fit = fitness(pop)
        best_scores.append(np.max(fit))

        if selection_strategy == 'tournament':
            selected = tournament_selection(pop, fit)
        elif selection_strategy == 'roulette':
            selected = roulette_wheel_selection(pop, fit)

        children = []
        for i in range(0, len(selected), 2):
            p1, p2 = selected[i], selected[(i+1)%len(selected)]
            if random.random() < crossover_rate:
                if crossover_strategy == 'one_point':
                    child = one_point_crossover(p1, p2)
                elif crossover_strategy == 'blend':
                    child = blend_crossover(p1, p2)
            else:
                child = p1
            children.append(child)

        pop = np.array([mutate(c, mutation_rate) for c in children])
        pop = np.clip(pop, 0, 2 * np.pi)

    return best_scores

# Run experiments
results = {
    "selection": {
        "tournament": genetic_algorithm('tournament', 'one_point'),
        "roulette": genetic_algorithm('roulette', 'one_point')
    },
    "crossover": {
        "one_point": genetic_algorithm('tournament', 'one_point'),
        "blend": genetic_algorithm('tournament', 'blend')
    },
    "mutation": {
        "0.1": genetic_algorithm('tournament', 'one_point', mutation_rate=0.1),
        "0.3": genetic_algorithm('tournament', 'one_point', mutation_rate=0.3),
        "0.5": genetic_algorithm('tournament', 'one_point', mutation_rate=0.5)
    }
}

# Plot
fig, axs = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

for i, (key, series) in enumerate(results.items()):
    for label, data in series.items():
        axs[i].plot(data, label=label)
    axs[i].set_title(f"Effect of {key.capitalize()}")
    axs[i].set_xlabel("Generation")
    axs[i].set_ylabel("Best Fitness")
    axs[i].legend()
    axs[i].grid(True)

fig.suptitle("Impact of Genetic Operators on sin(x) Optimization", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Apply Genetic Algorithms to solve a combinatorial optimization problem, such as the traveling salesman problem.
import numpy as np
import random
import matplotlib.pyplot as plt

# Target function
def true_sine(x):
    return np.sin(x)

# Fitness: Mean Squared Error between predicted and true sine values
def fitness(chromosome, x_vals):
    a, b, c = chromosome
    preds = a * np.sin(b * x_vals + c)
    return np.mean((true_sine(x_vals) - preds) ** 2)

# Initialize population
def init_population(size):
    return [np.random.uniform(-2, 2, size=3).tolist() for _ in range(size)]

# Selection (tournament)
def selection(pop, x_vals):
    candidates = random.sample(pop, 5)
    candidates.sort(key=lambda ch: fitness(ch, x_vals))
    return candidates[0]

# Crossover (blend)
def crossover(p1, p2):
    alpha = 0.5
    return [(1 - alpha) * p1[i] + alpha * p2[i] for i in range(3)]

# Mutation (Gaussian noise)
def mutate(chromosome, rate=0.1):
    return [gene + np.random.normal(0, rate) for gene in chromosome]

# Run GA
def run_ga(generations=100, pop_size=100):
    x_vals = np.linspace(0, 2 * np.pi, 100)
    population = init_population(pop_size)
    best = min(population, key=lambda ch: fitness(ch, x_vals))

    for gen in range(generations):
        new_pop = []
        for _ in range(pop_size):
            p1 = selection(population, x_vals)
            p2 = selection(population, x_vals)
            child = crossover(p1, p2)
            child = mutate(child)
            new_pop.append(child)

        population = new_pop
        current_best = min(population, key=lambda ch: fitness(ch, x_vals))
        if fitness(current_best, x_vals) < fitness(best, x_vals):
            best = current_best

        if gen % 10 == 0:
            print(f"Gen {gen} | Best MSE: {fitness(best, x_vals):.5f} | Params: {best}")

    return best, x_vals

# Run
best_params, x_vals = run_ga()
a, b, c = best_params
y_pred = a * np.sin(b * x_vals + c)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(x_vals, np.sin(x_vals), label="True sin(x)", linewidth=2)
plt.plot(x_vals, y_pred, label=f"GA Approx: {a:.2f}sin({b:.2f}x + {c:.2f})", linestyle='--')
plt.legend()
plt.grid(True)
plt.title("Genetic Algorithm Approximating Sine Curve")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Generate synthetic linearly separable data
np.random.seed(0)

# Class 1 (label = 1)
X1 = np.random.randn(10, 2) + np.array([2, 2])

# Class 0 (label = -1)
X2 = np.random.randn(10, 2) + np.array([-2, -2])
#print(X1)
#print(X2)
# Combine data
X = np.vstack((X1, X2))
y = np.hstack((np.ones(10), -np.ones(10)))  # Labels: 1 and -1
#print(X, y)

class Perceptron:
    def __init__(self, learning_rate=0.01, epochs=100):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        # Initialize weights and bias
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for _ in range(self.epochs):
            for i in range(X.shape[0]):
                update = self.lr * (y[i] - self.predict(X[i]))
                self.weights += update * X[i]
                self.bias += update

    def predict(self, X):
        return np.where(np.dot(X, self.weights) + self.bias >= 0, 1, -1)
# Train the perceptron
perceptron = Perceptron(learning_rate=0.1, epochs=10)
perceptron.fit(X, y)


# Plot data points
plt.scatter(X1[:, 0], X1[:, 1], color='blue', label='Class 1')
plt.scatter(X2[:, 0], X2[:, 1], color='red', label='Class 0')

# Plot decision boundary
x_vals = np.linspace(-5, 5, 100)
y_vals = -(perceptron.weights[0] * x_vals + perceptron.bias) / perceptron.weights[1]
plt.plot(x_vals, y_vals, 'g--', label='Decision Boundary')

plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Perceptron Decision Boundary')
plt.show()
test_point = np.array([-1.1, -1.2])
prediction = perceptron.predict(test_point)
print(f"Prediction for point {test_point}: {'Class 1' if prediction == 1 else 'Class 0'}")
```

```python
#Use Python libraries (e.g., TensorFlow/Keras) to train a neural network on the MNIST dataset for digit recognition
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
import matplotlib.pyplot as plt
import seaborn as sns

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize data to [0, 1] range
x_train, x_test = x_train / 255.0, x_test / 255.0

# Define MLP model
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Flatten 28x28 images to 1D
    Dense(128, activation='relu'),  # Hidden layer with 128 neurons
    Dense(10, activation='softmax')  # Output layer for 10 classes
])

# Compile model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train model
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Evaluate model
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")

# Plot loss curve
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Curve')
plt.legend()
plt.show()

# Sample prediction
sample_images = x_test[5:11]
predictions = np.argmax(model.predict(sample_images), axis=1)

# Display sample images with predictions
fig, axes = plt.subplots(1, 5, figsize=(10, 3))
for i, ax in enumerate(axes):
    ax.imshow(sample_images[i], cmap='gray')
    ax.set_title(f"Pred: {predictions[i]}")
    ax.axis('off')
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

# ---------------------- TRAINING THE MODEL ----------------------

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize data (scale pixel values to [0,1])
x_train, x_test = x_train / 255.0, x_test / 255.0

# Define Neural Network Model
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Convert 28x28 images to 1D
    Dense(128, activation='relu'),  # Hidden layer with 128 neurons
    Dense(10, activation='softmax')  # Output layer (10 classes for digits 0-9)
])

# Compile the Model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the Model
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Save the Model
model.save("mnist_digit_classifier.h5")  # Save model for later use

# Evaluate Model
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")

# Plot Training and Validation Loss
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Curve')
plt.legend()
plt.show()

# ---------------------- PREDICT FROM GOOGLE IMAGE ----------------------

# Function to download an image from a URL
def download_image(url, save_path="downloaded_digit.png"):
    response = requests.get(url)  # Fetch image from the internet
    if response.status_code == 200:
        img = Image.open(BytesIO(response.content))  # Convert response to an image
        img.save(save_path)  # Save the image locally
        print(f"Image downloaded and saved as {save_path}")
        return save_path
    else:
        print("Failed to download image.")
        return None

# Function to preprocess the downloaded image
def preprocess_image(image_path):
    img = Image.open(image_path).convert('L')  # Convert to grayscale
    img = img.resize((28, 28))  # Resize to 28x28 pixels
    img = np.array(img)  # Convert to NumPy array
    img = img / 255.0  # Normalize pixel values
    img = img.reshape(1, 28, 28)  # Reshape to match model input
    return img

# Function to predict the digit from an image
def predict_digit(image_url, model_path="mnist_digit_classifier.h5"):
    # Download image from URL
    image_path = download_image(image_url)
    if image_path is None:
        return

    # Load trained model
    model = tf.keras.models.load_model(model_path)

    # Preprocess image
    processed_img = preprocess_image(image_path)

    # Predict digit
    prediction = np.argmax(model.predict(processed_img), axis=1)

    # Display image and prediction
    plt.imshow(processed_img[0], cmap='gray')
    plt.title(f"Predicted Digit: {prediction[0]}")
    plt.axis('off')
    plt.show()

    print(f"The model predicts the digit as: {prediction[0]}")

# ---------------------- ADD IMAGE URL HERE ----------------------
image_url = "https://www.shutterstock.com/image-vector/digit-9-number-design-template-260nw-728861911.jpg"  # <-- PASTE IMAGE URL HERE

# Predict digit from online image
predict_digit(image_url)
```
